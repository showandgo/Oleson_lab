```python
import os
import folium
import geopandas as gpd
```

```python
c_maui_gdf = gpd.read_file("~/projects/asner19/data/maui_gpd.shp")
```

```python
c_maui_json = c_maui_gdf.to_json()    # Convert to json so folium can read Points
```

### The map below shows the Rec sites used in Carlo's model

```python
m = folium.Map(location=[20.7984, -156.3319])

points = folium.features.GeoJson(c_maui_json)
m.add_child(points)
m
```

```python
''' To create buffer around the points,
first reproject to epsg=7855, so can buffer in metres,
then reconvert back to WGS84 (epsg=4326) '''

buff_c_maui_gdf = c_maui_gdf.to_crs(epsg=7855).buffer(600).to_crs(epsg=4326)    # Buffer of 600m diameter
buff_c_maui_json = buff_c_maui_gdf.to_json()
```

### Map below shows points but with 300m (radius) buffer

```python
m = folium.Map(location=[20.7984, -156.3319])

points = folium.features.GeoJson(buff_c_maui_json)

m.add_child(points)
m
```
